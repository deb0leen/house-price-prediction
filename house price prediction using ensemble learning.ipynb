```python
anova_test_data = df_copy[~(df['SalePrice'].isnull())]
```

```python
from scipy.stats import f_oneway
```

```python
for col in anova_test_data.select_dtypes(include = 'object').columns:
    CategoryGroup = anova_test_data.groupby(col)['SalePrice'].apply(list)
    AnovaResults = f_oneway(*CategoryGroup)
    if AnovaResults[1] < 0.05:
        print('P-Value for Anova is: ', col, AnovaResults[1], 'Correlated')
    else:
        print('P-Value for Anova is: ', col, AnovaResults[1], 'Not Correlated')
```
